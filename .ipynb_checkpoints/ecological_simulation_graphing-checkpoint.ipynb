```python
import numpy as np

plots = 50
years = 35
species_num = 800
```

```python
def load_sim_results(filename):
    res = np.loadtxt('sim_results/' + filename)
    res = res.reshape((plots, years, species_num))
    return res
```

```python
def d0_div (sim_result, species_num, years, plots):
    d0_alpha_diversity = np.zeros(years)
    d0_beta_diversity = np.zeros(years)
    d0_gamma_diversity = np.zeros(years)
    # Calculating D0_diversities
    for y in range(years):
        # Calculating alpha diversity
        average_alpha_diversity = 0
        for p in range(plots):
            alpha_diversity = 0
            for s in range(species_num):
                if sim_result[p][y][s] > 0:
                    alpha_diversity += 1
            average_alpha_diversity = average_alpha_diversity + alpha_diversity
        d0_alpha_diversity[y] = average_alpha_diversity / plots

        # Calculating gamma_diversity
        gamma_diversity = 0
        for s in range(species_num):
            sp_boolean = False
            for p in range(plots):
                if sim_result[p][y][s] > 0:
                    sp_boolean = True
            if sp_boolean:
                gamma_diversity += 1
        d0_gamma_diversity[y] = gamma_diversity

        # Calculating beta diversity
        d0_beta_diversity[y] = d0_gamma_diversity[y] / d0_alpha_diversity[y]
    print("Calculated D0 Diversities")
    return (d0_alpha_diversity, d0_beta_diversity, d0_gamma_diversity)
```

```python
res1 = load_sim_results("temp700.csv")
res2 = load_sim_results("temp500.csv")
res3 = load_sim_results("temp400.csv")
```

```python
import seaborn
from matplotlib import pyplot as plt
```

```python
def plot_results(x_axis, y1, y2, y3, div_type, legend):
    fig = plt.figure()
    fig.suptitle('%s Diversities against Time' % div_type, fontsize=25, fontweight='bold')

    x = [i for i in range(years)]

    ax = fig.add_subplot(111)
    plt.plot(x, y1, color='r', linewidth=2.0)
    plt.plot(x, y2, color='b', linewidth=2.0)
    plt.plot(x, y3, color='g', linewidth=2.0)
    ax.set_xlabel('Years', fontsize=15)
    ax.set_ylabel('%s Diversities' % div_type, fontsize=15)
    ax.legend(legend)
    plt.show()
```

```python
x = [i for i in range(years)]
legend = ["TP=600", 'TP=500', 'TP=400']
plot_results(x, a1, a2, a3, 'Alpha', legend)
plot_results(x, b1, b2, b3, 'Beta', legend)
plot_results(x, g1, g2, g3, 'Gamma', legend)
```

```python
def calculate_d_zero(plots, species_global_pool, species_by_plot):
    d_zero = np.zeros((50,1))
    # for each plot
    for i in range(plots):
        # for each species
        for s in range(species_global_pool):
            # if the plot has that species, increment the count of species by one
            if species_by_plot[i][s] > 0:
                d_zero[i] += 1
                    
    return d_zero
```

```python
def calculate_gamma(species_matrix):
    
    # changes the species_matrix to a year by plot by species matrix
    modified = np.rollaxis(species_matrix, 1)
    
    # for each year
    for y in range(years):
        species_by_plot = modified[y]
        
        """calculating the Hill numbers:
             1) d_zero = count of species, not taking into account the abundance
             2) d_one = also known as Shannon's index: 
                                 given by p_i ln p_i, taking into account the proportion of abundances
             3) d_two = also known as Simpson's index:
                                 (to be researched!)
        """
        d_zero = calculate_d_zero(plots, species_global_pool, species_by_plot)
        # calculating alpha diversity
        alpha_bar = np.sum(d_zero) / plots
        print(alpha_bar)
        
calculate_gamma(species_matrix)
```
